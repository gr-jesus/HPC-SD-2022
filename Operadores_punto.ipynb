# Operaciones orientadas a punto

```python
import cv2
import time
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
original_image = cv2.imread('Lenna.png')
print(type(original_image))
```

```python
img_show = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
plt.imshow(img_show)
```

```python
def transform_image(intensidades, img_entrada):
    new_img=[]
    for fila in img_entrada:
        for columna in fila:
            for pixel in columna:
                new_img.append(intensidades[pixel])
                
    return np.array(new_img).astype(int).reshape(img_entrada.shape)
```

## Operador negativo

```python
vector_negativo=[]
for i in range(256):
    vector_negativo.append(255-i)
    
s_t=time.perf_counter()
img_negativa=transform_image(vector_negativo, img_show)
f_t=time.perf_counter()

print(f'Programa terminado en {f_t-s_t} segundos')

plt.imshow(img_negativa)
```
